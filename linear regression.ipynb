```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
height=np.array([5.2,5.7,5.9,6.0,6.4])
weight=np.array([125,140,155,160,180])
```

```python
plt.scatter(height, weight)
plt.xlabel("height (inches)")
plt.ylabel("weight (pounds)")
plt.show()
```

```python
m, b= np.polyfit(height, weight, 1)
```

```python
plt.scatter(height, weight)
plt.plot(height, m*height +b, color='red')
plt.xlabel("height (inches)")
plt.ylabel("weight (pounds)")
plt.show()
```

```python
new_height=5.8
predicted_weight=m*new_height+b
print(f"predicted weight: {predicted_weight}")
```

```python
#car purchasing prediction
import pandas as pd
from sklearn.linear_model import LinearRegression 
```

```python
file_path = 'Car_Purchasing_Data.csv'
for encoding in ['latin1', 'iso-8859-1', 'cp1252']:
    try:
        data = pd.read_csv(file_path, encoding=encoding)
        print(f"Successfully read the file with {encoding} encoding")
        break
    except Exception as e:
        print(f"Failed with {encoding} encoding: {e}")
data
```

```python
X=data[['Gender', 'Age', 'Annual Salary', 'Credit Card Debt', 'Net Worth']]
Y=data['Car Purchase Amount']
```

```python
model=LinearRegression()
model.fit(X,Y)  #training the model by fitting the data in it
```

```python
new_data=np.array([[1, 33, 75500, 15645, 3358]])
predicted_price=model.predict(new_data)
print(f"purchasing price will be: {predicted_price}")
```
